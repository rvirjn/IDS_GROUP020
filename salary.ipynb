The below code snippet is to normalize the data by removing the spaces, and replacing symbols with appropriate words. The subsequent code line does data cleansing activity by converting all the characters to lower case as the there would be no duplication with respect to data.

```python
data.columns = data.columns.str.replace(' ','_')
data.columns = data.columns.str.replace('#','no')
data.columns = data.columns.str.replace('/','_')
data.columns = data.columns.str.replace('-','_')
```

```python
 data.columns = data.columns.str.lower()
```

```python
data_modi = data
```

```python
data.job_id.dtype == 'O'
```

```python
data_modi = data.apply(lambda x: x.str.lower() if x.dtype == 'O' else x)
```

```python
pd.set_option('display.max_columns', None)
```

The below table shows data sample after conversion as per above code lines.

```python
data_modi.head()
```

The below code is performing data extraction from the mminimum qual requirements column to obtain the derived value of experience. The below table does the mapping and the 2nd table adds experience column to the data.

```python
data_modi.minimum_qual_requirements.value_counts()
```

```python
years = {'one': 1,
         'two':2,
        'three':3,
        'four':4,
        'five':5,
        'six':6,
        '4':4,
        'eight':8,
        'seven':7,
        'ten':10,
        '10':10,
        '3':3,
        '5':5,
        '2':2,
         'fifteen':15,
         '15':15,
}
```

```python
data_modi['experience'] = data_modi['minimum_qual_requirements'].str.extract(r'(\w+)\syears.+experience')
```

```python
data_modi['experience'] = data_modi['experience'].map(years)
```

```python
data_modi.head()
```

```python
data_modi['experience'] = data_modi.experience.apply(lambda x: 0 if np.isnan(x) else x)
```

```python
data_modi.experience.value_counts()
```

The below code is extracting just the required variable to perform the data pre-processing and to obtain required resutlts.


```python
data_req = data_modi.filter(['job_id', 'experience', 'job_category', 'salary_range_from', 'salary_range_to','preferred_skills'])
```

```python
data_req = data_req.dropna()
```

The below code snippet is to extract the preferred skils from the top 10 top skills.

```python
data_req['preferred_skills']
```

```python
data_req['job_category'] = data_req.job_category.str.replace('&','and')
```

```python
data_req
```

The below code snippet to obtain the top 10 paid salary for respective job skills. We are taking the mean of the salaries given and sorting it in the decreasing order.

```python
data_req.groupby('job_category').agg({'job_category':'count', 'salary_range_from': 'mean','salary_range_to': 'mean'})\
.sort_values(by = ['salary_range_to', 'salary_range_from'], ascending=False)
```

```python
data_req.groupby('job_category').agg({'job_category':'count', 'salary_range_from': ['min', 'mean'],'salary_range_to': ['max', 'mean']})\
.sort_values(by = [('job_category', 'count')], ascending=False)
```

The below code snippet gives the top 10 filtered data as per the requirement.

```python
data_req.groupby('job_category').agg({'job_category':'count', 'salary_range_from': ['min', 'mean'],
                                      'salary_range_to': ['max', 'mean'], 'experience': ['min', 'max', 'mean']})\
.sort_values(by = [('job_category', 'count')], ascending=False).head(10)
```

```python
data_req['job_category'].value_counts().head(10).plot(kind='bar')
```

```python
data_req.groupby('job_category').agg({'job_category':'count', 'salary_range_from': ['min', 'mean'],'salary_range_to': ['max', 'mean']})\
.sort_values(by = [('job_category', 'count')], ascending=False).boxplot()
```

```python
ax = data_req.groupby('job_category').head(10).plot(x="job_category", y="salary_range_from", kind="bar")
data_req.groupby('job_category').head(10).plot(x="job_category", y="salary_range_to", kind="bar", ax=ax, color="C2")
data_req.groupby('job_category').head(10).plot(x="job_category", y="experience", kind="bar", ax=ax, color="C3")
```
